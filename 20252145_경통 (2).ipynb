```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('SoftDrink.xlsx', sheet_name = 'Data',usecols = [0], header = 0)
```

```python
df
```

```python
Brands=df['Brand Purchased']
```

```python
brand, counts = np.unique(Brands, return_counts= True)
brand, counts
```

```python
freq = Brands.value_counts()
freq
```

```python
n = len(Brands)
n
```

```python
rel_freq =  freq / n
rel_freq
```

```python
pd.DataFrame({'freq' : freq, 'rel_freq ' : rel_freq})
```

```python
plt.bar(brand, counts)
```

```python
plt.bar(brand, counts)

plt.xlabel('Brands')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.pie(counts, labels=brand, autopct='%.1f%%')
plt.show()
```
