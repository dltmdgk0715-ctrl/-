```python
Audit = pd.read_excel('Audit.xlsx',usecols=[0],header = 0)
```

```python
audit_group = np.where(audit<=14, '10일~ 14일',
                       np.where(audit<=19, '15일~19일',
                                np.where(audit<=24,'20일~24일',
                                       np.where(audit<=29,'25일~29일','30일~34일'))))

```

```python
audit_counts = pd.Series(audit_group).value_counts()
```

```python
audit_table = pd.DataFrame({'freq' : audit_counts})
```

```python
audit_table.sort_index(inplace = True)
```

```python
audit_table.sort_index(inplace= True)
audit_table
```

```python
audit_table.loc['Sum'] = audit_table.sum(axis=0)
audit_table
```

```python
Restaurant = pd.read_excel('Restaurant.xlsx', header = 0 ,usecols = [1,2])
Restaurant
```

```python
Restaurant['Meal Price ($)'].min(),Restaurant['Meal Price ($)'].max()
```

```python
price_group = np.where(Restaurant['Meal Price ($)'] <= 19, '10~19',
                       np.where(Restaurant['Meal Price ($)'] <= 29, '20~29',
                       np.where(Restaurant['Meal Price ($)'] <= 39, '30~39','40~49')))

```

```python
price_group
```

```python
Restaurant['Group'] = price_group
Restaurant
```

```python
cross_table = pd.crosstab(index = Restaurant['Quality Rating'], columns = Restaurant['Group'])
cross_table
```

```python
Quality_order = ['Good', 'Very Good', 'Excellent']
```

```python
cross_table = cross_table.reindex(Quality_order)
cross_table
```

```python
cross_table['Sum'] = cross_table.sum(axis=1)
cross_table
```

```python
cross_table.loc['Sum'] = cross_table.sum(axis = 0)
cross_table
```

```python
joint_prob = cross_table / cross_table.loc['Sum','Sum']
joint_prob
```

```python
joint_prob.style.format('{:.4f}')
```

```python
row_percent = cross_table.div(cross_table['Sum'], axis=0) * 100
row_percent
```
