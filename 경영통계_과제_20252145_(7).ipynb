표본분포 개요

```python
# 기상의 데이터프레임 생성(총 2500개의 요소)
np.random.seed(1)

#30개짜리 표본 500개 추출, 각 표본에서 'Salary'의 평균, 'Training'의 Yes 비율 계산
#빈 자리
sample_means = []
sample_ratios = []

for _ in range(500):
  sample = df.sample(n = 30, random_state = np.random.randint(0, 10000))
  sample_mean = sample['Annual Salary'].mean()
  sample_means.append(sample_mean)
  sample_ratio = (sample['Training Program'] == 'Yes').mean()
  sample_ratios.append(sample_ratio)

# df_500sample (데이터프레임) 생성 : 500개의 표본평균과 표본비율로 구성
df_500sample = pd.DataFrame({
    'sample_mean':sample_means,
    'sample_ratio':sample_ratios
})
df_500sample
```
